Threshold

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


# "gazete.jpg" adlı görüntüyü siyah beyaz olarak oku
img = cv2.imread("gazete.jpg", 0)

# Adaptif eşikleme (ortalama) işlemi uygula
# img: İşlem yapılacak siyah beyaz görüntü
# 255: Maksimum piksel değeri
# cv2.ADAPTIVE_THRESH_MEAN_C: Adaptif eşikleme yöntemi (ortalama değeri kullan)
# cv2.THRESH_BINARY: Eşikleme sonucu ikili bir görüntü oluşturur
# 11: Blok boyutu - Her piksel için eşikleme değeri hesaplarken kullanılan komşu piksellerin boyutu
# 8: C sabiti - Ortalama değeri hesaplarken çıkarılacak sabit değer
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 8)

# Görüntüyü gösteren işlevi tanımla
def show_pic(img):
    fig = plt.figure(figsize=(15, 15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap="gray")

# Adaptif eşikleme sonucunu göster
show_pic(th2)



```

morphological işlemler

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Boş bir 600x600 piksel siyah bir görüntü oluşturan ve içine abcde yazan fonksiyon
def load():
    blank_img = np.zeros((600, 600))
    # Görüntüye "ABCDE" yazısı ekleyen işlem
    blank_img = cv2.putText(blank_img, 'ABCDE', (50, 300), cv2.FONT_HERSHEY_SIMPLEX, 5, (255, 255, 255), 30)
    return blank_img

# Görüntüyü ekranda gösteren işlev
def display_img(img):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')

# İlk görüntüyü oluştur
img = load()

# Aşındırma işlemi için bir çekirdek (kernel) oluştur
kernel = np.ones((5, 5), dtype=np.uint8)

# Görüntüyü erozyona uğrat (aşındırma) işlemi
img = cv2.erode(img, kernel=kernel, iterations=3)

# İkinci görüntüyü oluştur
img1 = load()

# Beyaz gürültü eklemek için rastgele beyaz gürültü oluştur
white_noise = np.random.randint(0, 2, (600, 600))
white_noise = white_noise * 255
img1 = img1 + white_noise

# Beyaz gürültüyü temizlemek için morfolojik açma işlemi uyguladık 
clear_img = cv2.morphologyEx(img1, cv2.MORPH_OPEN, kernel)

# Üçüncü görüntüyü oluştur
img3 = load()

# Siyah gürültü eklemek için rastgele siyah gürültü oluştur
black_noise = np.random.randint(0, 2, (600, 600))
black_noise = black_noise * -255
private = img3 + black_noise

# -255 değerlerini 0 yaparak siyah gürültüyü temizle
private[private == -255] = 0

# harflerin içindeki gürültüyü temizlemek için morfolojik kapatma işlemi uygula
private = cv2.morphologyEx(private, cv2.MORPH_CLOSE, kernel)

# cv2.MORPH_GRADIENT morfolojik işlemler içinde bir türdür bu işlem, genellikle kenarları belirlemek veya nesnelerin dış sınırlarını çıkarmak için kullanılır
private = cv2.morphologyEx(private, cv2.MORPH_GRADIENT, kernel)# 


# Oluşturulan son görüntüyü ekranda göster
display_img(private)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


img=cv2.imread("Sudoku.png")



# ekrana basmak için bir fonks
def display_img(img):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')


# 3 ve 4 parametreler x ve y eksenlerinde belirginleştrime işlemi yapıyıpor
sobel_y=cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5) # y ekseninde 
sobel_x=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)# x ekseninde


blended=cv2.addWeighted(sobel_x,0.5,sobel_y,0.5,0)# sobel x ve sobel y yi bland yaparak laplaciana yakın bir görüntü elde etmeye çalıştık


img=cv2.Laplacian(img,cv2.CV_64F)# x ve y eksenlerinde aynı anda belirginleştirme işlemi yapıyor



ret,th=cv2.threshold(img,50,255,cv2.THRESH_BINARY)


kernel=np.ones((4,4))# 4 e 4 lük kernel oluşturduk

mex=cv2.morphologyEx(blended,cv2.MORPH_GRADIENT,kernel)# rakamların içini beyaz ile doldurduk


display_img(mex)
```

```python
import cv2

# İki köşe noktasını saklamak için iki değişken tanımlayın
pt1 = (0, 0)
pt2 = (0, 0)

# Köşelerin seçilip seçilmediğini takip eden bayraklar
topLeft_clicked = False
topRight_clicked = False

# Dikdörtgen çizme işlevi
def dikdortgen_ciz(event, x, y, flags, param):
    global pt1, pt2, topLeft_clicked, topRight_clicked

    if event == cv2.EVENT_LBUTTONDOWN:
        if topLeft_clicked and topRight_clicked:
            pt1 = (0, 0)
            pt2 = (0, 0)
            topLeft_clicked = False
            topRight_clicked = False

        if not topLeft_clicked:
            pt1 = (x, y)
            topLeft_clicked = True

        elif not topRight_clicked:
            pt2 = (x, y)
            topRight_clicked = True

# Kamera yakalaması oluşturun
cap = cv2.VideoCapture(0)

# Bir pencere oluşturun ve fare olaylarını izlemek için işlevi atayın
cv2.namedWindow('test')
cv2.setMouseCallback('test', dikdortgen_ciz)

while True:
    ret, frame = cap.read()
    frame = cv2.flip(frame, 1)

    # Birinci köşeyi işaretleyin
    if topLeft_clicked:
        cv2.circle(frame, pt1, 5, (0, 0, 255), -1)

    # İkinci köşeyi işaretleyin ve dikdörtgeni çizin
    if topLeft_clicked and topRight_clicked:
        cv2.rectangle(frame, pt1, pt2, (0, 0, 255), 2)

    cv2.imshow('test', frame)

    if cv2.waitKey(5) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#  bu fonksiyon görüntüleri çizdirmek için kullanılacak
def display(img, cmap="gray"):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap=cmap)

# İlk görüntüyü "gevrek.png" olarak yükle (gri tonlamalı)
urun = cv2.imread("gevrek.png", 0)

# İkinci görüntüyü "raf.jpg" olarak yükle (gri tonlamalı)
raf = cv2.imread("raf.jpg", 0)

# ORB (Oriented FAST and Rotated BRIEF) özellik tanıma nesnesini oluştur
orb = cv2.ORB_create()

# İlk görüntüdeki ORB özelliklerini tespit et ve açıklamalarını hesapla
kp1, dest1 = orb.detectAndCompute(urun, None)

# İkinci görüntüdeki ORB özelliklerini tespit et ve açıklamalarını hesapla
kp2, dest2 = orb.detectAndCompute(raf, None)

# BFMatcher ile ORB özelliklerini eşleştir
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)

# Eşleşen özellikleri bul
matches = bf.match(dest1, dest2)

# Eşleşenleri uzaklıklarına göre sırala
matches = sorted(matches, key=lambda x: x.distance)

# İlk 25 eşleşmeyi görsel olarak işaretle
cizdirme = cv2.drawMatches(urun, kp1, raf, kp2, matches[:25], None, flags=2)

# Sonucu görselleştir
display(cizdirme)
```

whatershed algoritması

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# display() adlı bir fonksiyon tanımlanıyor, bu fonksiyon görüntüleri çizdirmek için kullanılacak
def display(img, cmap="gray"):
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap=cmap)

# "coin.jpg" adlı resmi yükleyin
para = cv2.imread("coin.jpg")

# Resmi median bulanıklaştırma işlemi ile işleyin
img = cv2.medianBlur(para, 5)

# Gri tonlamalı bir görüntüye dönüştürün
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Bir eşikleme işlemi uygulayın
ret, thresh = cv2.threshold(img, 160, 255, cv2.THRESH_BINARY_INV)

# Görüntüdeki konturları bulun
conturs, _ = cv2.findContours(thresh.copy(), cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)

# Konturları işaretle
for i in range(len(conturs)):
    if i < 1:
        cv2.drawContours(para, conturs, i, (255, 0, 0), 2)

# Bir morfolojik işlem (kapama) uygulayın
kernel = np.ones((3, 3), np.uint8)
opening = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel, 2)

# Genişletme işlemi uygulayın
dilate = cv2.dilate(opening, kernel, 3)

# Uzaklık dönüşümünü uygulayın
dist_transform = cv2.distanceTransform(dilate, cv2.DIST_L2, 5)

# Bir eşikleme işlemi daha uygulayın
ret2, thresh2 = cv2.threshold(dist_transform, 0.7 * dist_transform.max(), 255, 0)

# Eşiklenmiş görüntüyü tam sayıya dönüştürün
thresh2 = np.uint8(thresh2)

# Bilinmeyen bölgeleri belirleyin
unknow = cv2.subtract(thresh, thresh2)

# Etiketleme işlemi uygulayın
ret, markers = cv2.connectedComponents(thresh2)
markers = markers + 1
markers[unknow == 255] = 0

# Watershed algoritması uygulayın
markers = cv2.watershed(para, markers)

# Yeniden konturları bulun
conturs, _ = cv2.findContours(markers.copy(), cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)

# Konturları işaretle
for i in range(len(conturs)):
    cv2.drawContours(para, conturs, i, (255, 0, 0), 1)

# Sonucu görselleştir
display(para)
```

optical flow

```python
import numpy as np
import cv2

# Köşe takip parametreleri
corner_track_params = dict(maxCorners=10,        # Takip edilecek köşe sayısı
                          qualityLevel=0.3,      # Köşelerin kalitesi
                          minDistance=7,         # Köşeler arasındaki minimum mesafe
                          blockSize=7)           # Köşeleri bulmak için kullanılan blok boyutu

# Optik akış hesaplama parametreleri
lk_params = dict(winSize=(200, 200),                # Optik akış için pencere boyutu
                 maxLevel=2,                        # Piramit seviyeleri
                 criteria=(cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 0.03))  # Hesaplama kriterleri

# Kamera yakalama
cap = cv2.VideoCapture(0)  # 0, varsayılan kamera cihazını temsil eder. Gerektiğinde değiştirilebilir.

# Başlangıç karesini yakala
ret, prev_frame = cap.read()
prev_gray = cv2.cvtColor(prev_frame, cv2.COLOR_BGR2GRAY)

# Başlangıç köşelerini hesapla
prev_pts = cv2.goodFeaturesToTrack(prev_gray, mask=None, **corner_track_params)

# Maske oluştur
mask = np.zeros_like(prev_frame)

# Ana döngü
while True:
    ret, frame = cap.read()  # Yeni bir kare yakala
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)  # Kareyi gri tona dönüştür

    # Optik akışı hesapla
    next_pts, status, err = cv2.calcOpticalFlowPyrLK(prev_gray, gray, prev_pts, None, **lk_params)

    # İyi takip edilen yeni ve önceki köşeleri seç
    good_new = next_pts[status == 1]
    good_prev = prev_pts[status == 1]

    # Köşeler arasında çizgi çek ve halka çiz
    # İki liste üzerinde aynı indeksle işlem yapmak için enumerate ve zip kullanma
    for i, (new, prev) in enumerate(zip(good_new, good_prev)):
        #ravel çok boyutlu bir diziyi tek boyutlu bir diziye dönüştürür
        x_new, y_new = new.ravel()
        x_prev, y_prev = prev.ravel()
        pt1 = (int(x_new), int(y_new))
        pt2 = (int(x_prev), int(y_prev))
        mask = cv2.line(mask, pt1, pt2, (0, 255, 0), 3)  # Köşeler arasında çizgi çek
        frame = cv2.circle(frame, pt1, 8, (0, 0, 255), -1)  # Köşenin etrafına halka çiz

    img = cv2.add(frame, mask)  # Çerçeveye maskeyi ekleyin
    cv2.imshow("frame", img)  # Sonucu göster

    if cv2.waitKey(10) == 27:  # ESC tuşuna basılınca döngüyü kapat
        break

    prev_gray = gray.copy()  # Önceki gri kareyi güncelle
    prev_pts = good_new.reshape(-1, 1, 2)  # Önceki köşeleri güncelle

cap.release()  # Kamerayı serbest bırak
cv2.destroyAllWindows()  # Pencereyi kapat







```

```python
import numpy as np
import cv2

# Kameradan video yakalamak için bir VideoCapture nesnesi oluşturun (0, varsayılan kamera cihazını temsil eder).
cap = cv2.VideoCapture(0)

# İlk kareyi yakala ve bu kareyi frame1'e atayın.
ret, frame1 = cap.read()

# İlk kareyi siyah-beyaz bir görüntüye dönüştürün (optik akış işlemi için).
prvsImg = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)

# Hareketi göstermek için renkli bir maske oluşturun.
hsv_mask = np.zeros_like(frame1)
hsv_mask[:, :, 1] = 255  # Parlaklık kanalını ayarla.

while True:
    # Yeni bir kareyi yakala ve bu kareyi frame2'ye atayın.
    ret, frame2 = cap.read()

    # Yeni kareyi siyah-beyaz bir görüntüye dönüştürün (optik akış işlemi için).
    nextImg = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)

    # Optik akışı hesaplayın (Farneback yöntemi).
    flow = cv2.calcOpticalFlowFarneback(prvsImg, nextImg, None, 0.5, 3, 15, 3, 5, 1.2, 0)

    # Hız ve yöne dair bilgileri alın.
    mag, ang = cv2.cartToPolar(flow[:, :, 0], flow[:, :, 1], angleInDegrees=True)

    # Hareketi görselleştirmek için renkli maskeyi güncelleyin.
    hsv_mask[:, :, 0] = ang / 2
    hsv_mask[:, :, 2] = cv2.normalize(mag, None, 0, 255, cv2.NORM_MINMAX)

    # HSV formatındaki maskeyi BGR renk formatına dönüştürün (görüntüyü oluşturun).
    bgr = cv2.cvtColor(hsv_mask, cv2.COLOR_HSV2BGR)

    # Hareketi gösteren görüntüyü gösterin.
    cv2.imshow("frame2", bgr)
    
    # ESC tuşuna basıldığında döngüyü kırın.
    if cv2.waitKey(30) == 27:
        break

    # Önceki kareyi güncelleyin.
    prvsImg = nextImg

# Kamerayı serbest bırakın ve pencereyi kapatın.
cap.release()
cv2.destroyAllWindows()
```
